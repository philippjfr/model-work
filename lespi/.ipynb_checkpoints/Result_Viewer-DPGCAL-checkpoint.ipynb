```python
def update_files():
    plots = ["Orientation","Phase","Activity"]
    times = []
    sheets = []
    projections = []
    png_filepaths = [path for path in glob.glob('*.png') if path[0].isdigit()]
    for png in png_filepaths: 
        png_info = png[:-4].split('_')
        tid = png_info[2]
        time = png_info[4]
        if time not in times and len(time) == 9: times.append(time)
        sheet = png_info[5]
        if sheet not in sheets: sheets.append(sheet)
        if np.all([png_info[6]==plot for plot in plots]):
            projection = png_info[6]
            if projection not in projections: projections.append(projection)
    date = png_info[0]
    model = png_info[1]
    
    return sorted(times), sheets, projections
```
